##Q4. Write a Python program which accepts the radius of a circle from the user and compute the area.

```python
pi= 3.14159
r= float(input("enter radius:"))
Area_Circle= pi*(r**2)
if r<=0:
    print("Area is 0")
else:
    print("Area is:", Area_Circle)
```
